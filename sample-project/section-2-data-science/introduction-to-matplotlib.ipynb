```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Plotting : 1st method
plt.plot()
```

```python
# puting axesis     
x = [1,2,3,4,5]  # The axes process only arrays
y = [10, 20, 30, 40, 50]
plt.plot(x, y);
```

```python
# 2st method
fig = plt.figure() # Creates a figure
ax = fig.add_subplot() # Add the axes
plt.show()
```

```python
# 3st figure 
fig = plt.figure()
ax = fig.add_axes([1,1,1,1]) # Again; add_axes process only arrays
```

```python
# 4st method : Recommended (Why?)

fig, ax = plt.subplots()
ax.plot(x, [50, 100, 200, 250, 300])
```

### Matplotlib Example workflow

```python
# 1. Prepare data
x = [1,2,3]
y = [1,2,3]


# 2. Create plot 
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Put data inside plot
ax.plot(x, y)

# 4. Customize plot

ax.set(title="Simple plot",
       xlabel='x-axis')

# 5. Save the fig
fig.savefig("../images/sample-plot.png")
```

### Making figures with NP arrays

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
x[:10]
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**8)
```

```python
# Creatting scatter plot

fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make plot from dictionary 

nut_butter_prices = {
    "Almond" : 10,
    "Peanut" : 20,
    "Cashew" : 30
}

fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values())
ax.set(title="Nut butter prices",
      ylabel="Price $",)
```
